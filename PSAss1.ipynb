```python
def bracelet(ab, memory=None, depth=0, max_depth=100):
    # Start the list called memory
    if memory is None:
        memory = []

    # Turn the input into two single digit numbers
    y = ab // 10
    x = ab % 10
    
    # Calculate the next term in the sequence
    c = x
    d = (y + x) % 10
    cd = c * 10 + d
    
    # Add cd to memory
    memory.append(cd)
    
    # Stop the function when cd is equal to the original output
    if cd == ab or depth >= max_depth:
        return memory
    else:
        # Recursively call the function with cd as the new input
        return bracelet(cd, memory, depth + 1, max_depth)
bracelet(15)
```
